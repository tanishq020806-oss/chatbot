```python
def reply(user):
    if user in("hello","hi","hey"):
        return "Hi!"
    elif user in("how are you","how are you?","How are you","How are you?"):
        return "I’m good, thanks for asking. How about you?"
    elif user in("who are you?","what are you?","what's your name","who are you","what are you"):
        return "i'm simply a Chat Bot"
    elif user in("Bye","bye","GoodBye","see you"):
        return "GOOODBYE"

    if user=="":
        return "you didn't type anything! say something.."
    else:
        return "Sorry! I dont understand please try something like :'hello','how are you','Bye'"

def chat():
    print("Chatbot is ready! Type 'bye' or 'exit' to stop.\n")

    while True:
        user = input("You: ")
        response = reply(user)
        print("Bot:", response)
chat()
    
    
```
